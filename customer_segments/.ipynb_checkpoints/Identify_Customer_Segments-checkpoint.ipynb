```python
# function to convert feat_info.missing_or_unknown strings into list type
def string_to_list(s):
    """
    Convert item from string to list type
    Args: s - string
    Returns: s - list
    """
    
    if s == '[-1,X]':
        return [-1,'X']
    elif s == '[XX]':
        return ['XX']
    elif s == '[-1,XX]':
        return [-1,'XX']
    else:
        return eval(s)
```

```python
# convert value type for feat_info.missing_or_unknown
feat_info.missing_or_unknown = feat_info.missing_or_unknown.apply(string_to_list)
```

```python
# Create a dictionary to map column names to missing data encodings
missing_data_encode = dict(zip(feat_info.attribute, feat_info.missing_or_unknown))
```

```python
# function to one hot encode categorical features
def one_hot(df, col_name):
    dummies = pd.get_dummies(df[col_name])
    dummies.columns = [col_name + str(s) for s in dummies.columns]
    df.drop(col_name, inplace=True, axis=1)
    df = pd.concat([df, dummies], axis=1)
    return df
```

```python
# Investigate "PRAEGENDE_JUGENDJAHRE" and engineer two new variables.

def make_interval(x):
    
    if x in (1,2):
        return 1
    elif x in (3,4):
        return 2
    elif x in (5,6,7):
        return 3
    elif x in (8,9):
        return 4
    elif x in (10,11,12,13):
        return 5
    elif x in (14,15):
        return 6
    
def make_movement(x):
    
    if x in (2,4,6,7,9,11,13,15):
        return 0
    elif x in (1,3,5,8,10,12,14):
        return 1
```

```python
df_new['decade'] = df_new['PRAEGENDE_JUGENDJAHRE'].apply(make_interval)
df_new['movement'] = df_new['PRAEGENDE_JUGENDJAHRE'].apply(make_movement)
```

```python
# Investigate "CAMEO_INTL_2015" and engineer two new variables.

def wealth(x):
    
    if x // 10 == 1:
        return 1
    elif x // 10 == 2:
        return 2
    elif x // 10 == 3:
        return 3
    elif x // 10 == 4:
        return 4
    elif x // 10 == 5:
        return 5
    
def life_stage(x):
    
    if x % 10 == 1:
        return 1
    elif x % 10 == 2:
        return 2
    elif x % 10 == 3:
        return 3
    elif x % 10 == 4:
        return 4
    elif x % 10 == 5:
        return 5
```

```python
def get_digit(num,digit):
    if not pd.isnull(num):
        num = int(str(num)[digit])
    return num

df_new['wealth'] = df_new['CAMEO_INTL_2015'].apply(get_digit, digit=0)
df_new['life_stage'] = df_new['CAMEO_INTL_2015'].apply(get_digit, digit=1)
```

Then, when it comes to looking at the customer data in Step 3, you can just run this function on that DataFrame to get the trimmed dataset in a single step.

```python
def clean_data(df,missing_data_encode,col_remove_highmissing,feat_info):
    """Perform feature trimming, re-encoding, and engineering for demographics data
    
    Args:
    df: data frame. The data frame that need to be cleaned 
    missing_data_encode: dictionary. Key:column name. Value: missing value representor
    col_remove_highmissing: list. Columns with high missing rate that need to be removed
    feat_info: data frame. Feature info data.
    
    Returns:
    df: data frame. Cleaned data frame.
    df_highna: data frame. Data that has high na row values.
    """
    
    # Put in code here to execute all main cleaning steps:
    # convert missing value codes into NaNs, ...
    df = missing_code_to_nan(df, missing_data_encode)
    print('Done processing missing values from strings to nan')
    
    # remove selected columns and rows, ...
    df = drop_high_missing_col(df, col_remove_highmissing)
    df, df_high_na = drop_high_missing_row(df, threshhold = 0.1)
    print('Done removing high missing values in columns and rows')
    
    # select, re-encode, and engineer column values.
    df = re_encode_categorical_feature(df, feat_info)
    df = engineer_mixed_feature(df, feat_info)
    print('Done re-encoding and engineering features')
    
    # Return the cleaned dataframe.
    return df, df_high_na
```

```python
def missing_code_to_nan(df, missing_data_encode):
    for col in missing_data_encode.keys():
        # get row index for missing values in col
        missing_values_ridx = df[col].isin(missing_data_encode[col])
        # set missing values to nan
        df.loc[missing_values_ridx, col] = np.nan    
    return df


def drop_high_missing_col(df, col_remove):
    return df.drop(col_remove, axis=1)


def drop_high_missing_row(df, threshhold = 0.1):
    df['row_miss_per'] = df.isnull().mean(axis=1)
    df_lowna = df.loc[df['row_miss_per'] < threshhold].drop(['row_miss_per'], axis=1)
    df_highna = df.loc[df['row_miss_per'] > threshhold].drop(['row_miss_per'], axis=1)
    return df_lowna, df_highna


def re_encode_categorical_feature(df,feat_info):
    # change binary feature OST_WEST_KZ to numeric
    df['OST_WEST_KZ'] = df['OST_WEST_KZ'].map({'O':0,'W':1})
    
    # one hot multi-level features
    col_cate_num = df[feat_info[feat_info.type=='categorical'].attribute].nunique().sort_values()
    col_cate_num = col_cate_num.to_frame()
    col_cate_num.reset_index(level=0, inplace=True)
    col_cate_num.columns = ['colname','uniquen']
    
    muti_cate_col = col_cate_num[col_cate_num.uniquen>2].colname
    for cname in muti_cate_col:
        df = one_hot(df, cname)
    return df


def engineer_mixed_feature(df,feat_info):
    # engineer two features
    df = engineer_PRAEGENDE_JUGENDJAHRE(df)
    df = engineer_CAMEO_INTL_2015(df)
    # drop mixed features
    df = df.drop(feat_info[feat_info.type=='mixed'].attribute, axis=1)
    return df

def engineer_PRAEGENDE_JUGENDJAHRE(df):
    decade_map ={1: 1, 2: 1, 
             3: 2, 4: 2, 
             5: 3, 6: 3, 7:  3,
             8: 4, 9: 4, 
             10: 5, 11: 5, 12: 5, 13: 5,
             14: 6, 15: 6}
    
    movement_map = {1: 1, 3: 1, 5: 1, 8: 1, 10: 1, 12: 1, 14: 1,
                2: 2, 4: 2, 6: 2, 7: 2, 9: 2, 11: 2, 13: 2, 15: 2}
    
    df['decade'] = df['PRAEGENDE_JUGENDJAHRE'].map(decade_map)
    df['movement'] = df['PRAEGENDE_JUGENDJAHRE'].map(movement_map)
    return df

def get_digit(num, digit):
    if not pd.isnull(num):
        num = int(str(num)[digit])
    return num

def engineer_CAMEO_INTL_2015(df):
    df['wealth'] = df['CAMEO_INTL_2015'].apply(get_digit, digit=0)
    df['whealth'] = df['CAMEO_INTL_2015'].apply(get_digit, digit=1)
    return df
```

```python
def scree_plot(pca):
    
    '''
    Creates a scree plot associated with the principal components 
    
    INPUT: pca - the result of instantian of PCA in scikit learn
            
    OUTPUT:
            None
    '''
    components_total = len(pca.explained_variance_ratio_)
    components = np.arange(components_total)
    
    fig, ax1 = plt.subplots(figsize=(15,5))

    ax1.bar(components,pca.explained_variance_ratio_)
    ax1.set_ylabel('Explained Variance')
    ax1.set_xlabel('Number of Components')

    ax2 = ax1.twinx() 
    ax2.plot(np.cumsum(pca.explained_variance_ratio_), color="blue")
    ax2.set_ylabel('Cumulative Explained Variance')
    plt.title("Cumulative Explained Variance vs No. of Principal Components")
    plt.show()
    
```

### Step 2.3: Interpret Principal Components

Now that we have our transformed principal components, it's a nice idea to check out the weight of each variable on the first few components to see if they can be interpreted in some fashion.

As a reminder, each principal component is a unit vector that points in the direction of highest variance (after accounting for the variance captured by earlier principal components). The further a weight is from zero, the more the principal component is in the direction of the corresponding feature. If two features have large weights of the same sign (both positive or both negative), then increases in one tend expect to be associated with increases in the other. To contrast, features with different signs can be expected to show a negative correlation: increases in one variable should result in a decrease in the other.

- To investigate the features, you should map each weight to their corresponding feature name, then sort the features according to weight. The most interesting features for each principal component, then, will be those at the beginning and end of the sorted list. Use the data dictionary document to help you understand these most prominent features, their relationships, and what a positive or negative value on the principal component might indicate.
- You should investigate and interpret feature associations from the first three principal components in this substep. To help facilitate this, you should write a function that you can call at any time to print the sorted list of feature weights, for the *i*-th principal component. This might come in handy in the next step of the project, when you interpret the tendencies of the discovered clusters.

```python
def pca_weights(pca, i):
    
    weight_map = {}
    
    for counter, feature in enumerate(df_scaled.columns):
        weight_map[feature] = pca.components_[i][counter]
    
    sorted_weights = sorted(weight_map.items(), key=operator.itemgetter(1), reverse=True)
    
    return sorted_weights
```

- Then, compute the average difference from each point to its assigned cluster's center. **Hint**: The KMeans object's `.score()` method might be useful here, but note that in sklearn, scores tend to be defined so that larger is better. Use this fact to select a final number of clusters in which to group the data. **Warning**: because of the large size of the dataset, it can take a long time for the algorithm to resolve. The more clusters to fit, the longer the algorithm will take. You should test for cluster counts through at least 10 clusters to get the full picture, but you shouldn't need to test for a number of clusters above about 30.
- Once you've selected a final number of clusters to use, re-fit a KMeans instance to perform the clustering operation. Make sure that you also obtain the cluster assignments for the general demographics data, since you'll be using them in the final Step 3.3.

```python
def get_kmeans_score(data, k_value):
    
    '''
    returns the kmeans score regarding SSE for points to centers
    
    INPUT:
        data - the dataset you want to fit kmeans to
        k_value - the number of centers you want (the k value)
        
    OUTPUT:
        score - the SSE score for the kmeans model fit to the data
    '''

    # instantiate k-means
    kmeans = KMeans(n_clusters = k_value)

    # fit the model to your data using the fit method
    model = kmeans.fit(data)
    
    # obtain a score related to the model fit
    score = np.abs(model.score(data))
    
    return score
```

```python
# Over a number of different cluster counts...
# run k-means clustering on the data and...
# compute the average within-cluster distances.

scores = []
centers = list(range(1, 20, 2))

for center in centers:
    scores.append(get_kmeans_score(df_pca_50, center))
```

```python
# better understanding of clusters.
def map_cluster_to_features(kmeans, df, cnum):
    
    '''Map pca weights to  individual features
    and return two pd.Series on with the highest
    positive weights and one with the lowest negative
    weights
    
    Args:
    kmeans: kmeans object.
    df: data frame. The pca data frame of original data.
    cnum: int. The n-th cluster.
    
    Returns:
    top_pca_features: data frame. The top 'n' positive and negative pca features. 
    '''
    
    # get coordinates of cluster centers
    weights = pd.DataFrame(np.round(kmeans.cluster_centers_, 4), columns=df.keys())
    
    # the centroid of cluster cnum
    centroid = weights.iloc[cnum, :]
    
    # postive pca feature
    cent_pos = centroid[centroid > 0].sort_values(ascending=False).to_frame()
    cent_pos.reset_index(level=0, inplace=True)
    cent_pos.columns = ['top_pos_col','top_pos_col_weight']
    
    # negative pca feature
    cent_neg = centroid[centroid < 0].sort_values(ascending=True).to_frame()
    cent_neg.reset_index(level=0, inplace=True)
    cent_neg.columns = ['top_neg_col','top_neg_col_weight']
    
    # concat postive and negative
    top_pca_features = pd.concat([cent_pos, cent_neg], axis=1)
    top_pca_features['top_neg_col'] = top_pca_features['top_neg_col'].apply(str)
    top_pca_features['top_pos_col'] = top_pca_features['top_pos_col'].apply(str)
    
    return top_pca_features
```

### Discussion 3.3: Compare Customer Data to Demographics Data

**People who are likely to be a good target audience for the company**

Looking at the analysis, we can clearly notice that cluster 2 and 8 are popular with the company compared to the general population. In cluster 2 and 8, PCA components 0 and 1 as well as 6 and 4 have the highest weights - they relate to people with low income leaving in high density populated areas. These people are more likely to be the users of the mail-order company.

**People who are likely to not be a target audience for the company**

On the other hand, cluster 6 and 9 are rather unpopular with the company. Here, PCA components 7 and 3 have high positive weight, while components 1 and 5 have negative weight. High earners, living in outer areas and being more combative, dominant and cultural minded tend to not be a good target for the company.
